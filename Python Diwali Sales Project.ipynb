### EDA(Exploratory Data Analysis)

### Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# now checking which femal and male spend maximum amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen,hue='Gender')
plt.title('Amount Spend as per Gender',color='red')
plt.tight_layout()
plt.show()

for bars in ax.containers:
    ax.bar_label(bars)
```

## Insights and Observation

From above graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
# Total Amount vs Total Age

sales_age = df.groupby(['Age Group','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age,hue='Gender')

```

### State

```python
df.columns
```

```python
df['State'].unique()
```

```python
# Total number of orders from 10 top state

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sales_state.style.background_gradient(cmap='orange')
fig, ax = plt.subplots(figsize=(15, 5))
sns.barplot(x='State', y='Orders', data=sales_state, ax=ax)
plt.title('Number of orders from top 10 states')
plt.show()
```

```python
# Total amount/sales from top 10 states

sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(18, 5))
sns.barplot(x='State', y='Amount', color='purple',data=sales_state, ax=ax)
plt.title('Number of Amount from top 10 states',color='red')
plt.show()
```

### Insights and Observation from State

From above graphs we can see that most of the orders and total sales/amount are from Uttar Pradesh,Maharastra and Kernataka respectively

### Product Category

```python
df.columns
```

```python
df['Product_Category'].unique()
```

```python
# finding count based on Product category

sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(x='Product_Category',data=df,palette='viridis')
plt.title('Product Category Count',color='red')
plt.tight_layout()

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# checking amount spend based on product category

sales_amount=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='Product_Category', y='Amount', data=sales_amount, palette='viridis',legend=False,hue='Product_Category')
plt.title('Amount Spend as per Product Category',color='red')
plt.tight_layout()
plt.show()
```

### Insights and Observation from Product Category

From the above graphs we can see that most of orders done from Clothing and Apparel.While most of the amount spend on order done for Food product category
